```python
wiki_title = "Boston"
```

```python
from pathlib import Path

import requests


response = requests.get(
    "https://en.wikipedia.org/w/api.php",
    params={
        "action": "query",
        "format": "json",
        "titles": wiki_title,
        "prop": "extracts",
        # 'exintro': True,
        "explaintext": True,
    },
).json()
page = next(iter(response["query"]["pages"].values()))
wiki_text = page["extract"]

data_path = Path("data")
if not data_path.exists():
    Path.mkdir(data_path)

with open(data_path / f"{wiki_title}.txt", "w") as fp:
    fp.write(wiki_text)
```

```python
from llama_index import SimpleDirectoryReader, ListIndex

wiki_title = "Michael Jordan"
docs = SimpleDirectoryReader(
    input_files=[f"data/{wiki_title}.txt"]
).load_data()
```

```python
from llama_index.llms import OpenAI
from llama_index import ServiceContext

# works
# llm = OpenAI('text-davinci-003')
# doesn't work
llm = OpenAI('gpt-3.5-turbo-0613')
service_context = ServiceContext.from_defaults(llm=llm)
```

```python
list_index = ListIndex.from_documents(docs, service_context=service_context)

query_engine = list_index.as_query_engine()
response = query_engine.query(f"Give me a summary of {wiki_title}")
```
